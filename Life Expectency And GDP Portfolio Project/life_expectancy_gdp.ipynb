```python
from pyforest import *
```

```python
df= pd.read_csv('all_data.csv')
df.columns
```

```python

from warnings import simplefilter
simplefilter('ignore')
fig, axs = plt.subplots(nrows=2, ncols=3, sharex=True, sharey=True, figsize=(16,8))
axs= axs.flatten()
for i ,cty in enumerate(df.Country.unique()) :
    sns.barplot(df.Year,df[df.Country==cty]['Life expectancy at birth (years)'],ax=axs[i])
    axs[i].set_xticks(range(len(set(df.Year))))
    axs[i].set_xticklabels(list(set(df.Year)),rotation=45)
    axs[i].set_xlabel('Year')
    axs[i].set_ylabel('Life expectancy at birth (years)')
    axs[i].set_title(f"Year vs Life expectancy at birth (years) for  {cty}")
plt.show()
```

```python
fig,axs = plt.subplots(2,3, sharex=True, sharey=True, figsize=(14,7))
axs = axs.flatten()
Years = df.Year.unique()
Countries = df.Country.unique()
sns.set_palette("Paired")
for i,cty in enumerate(Countries):
    sns.barplot(x = "Year", y = "Life expectancy at birth (years)",
                data = df[df.Country==cty],ax=axs[i])
    axs[i].set_xticks(range(len(Years)))
    axs[i].set_xticklabels(Years,rotation=90)
    plt.xlabel('Year')
    plt.ylabel('Life expectancy at birth (years)')
    
plt.show()  

```
